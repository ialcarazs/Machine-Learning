# PREDICT ACCEPTIBITITY OF A CAR

# DISEO DEL PROYECTO

Este conjunto de datos de nivel principiante se deriv de un modelo de toma de decisiones que se desarroll originalmente para la investigacin sobre la toma de decisiones de atributos mltiples. La toma de decisiones implica la seleccin entre alternativas aparentemente conflictivas.

El conjunto de datos tiene 1728 filas y 7 columnas en las que los atributos del automvil, como el precio y la tecnologa, se describen en 6 atributos como "Precio de compra", "Mantenimiento" y "Seguridad", etc. Hay varias alternativas en cada uno de los 6 atributos. . La aceptabilidad del automvil, el sptimo atributo, es la variable de resultado.         


## OBJETIVO

Predecir la aceptabilidad de un coche.


Este conjunto de datos se recomienda para aprender y practicar sus habilidades en tcnicas de modelado de clasificacin .
