# Can you estimate the location from WIFI Signal Strength

# DISEO DEL PROYECTO

Este conjunto de datos de nivel principiante tiene 2000 filas y 8 columnas. El conjunto de datos contiene la intensidad de la seal wifi observada desde 7 dispositivos wifi en un telfono inteligente recopilados en un espacio interior. Los datos podran usarse para estimar la ubicacin en una de las cuatro habitaciones.

- Recommended Use: Classification Models
- Domain: Mobile/Location

## OBJETIVO

Predecir la ubicacin a partir de la intensidad de la seal wifi.

Este conjunto de datos se recomienda para aprender y practicar sus habilidades en anlisis de datos exploratorios , visualizacin de datos y tcnicas de modelado de clasificacin .
